```python
import pandas as pd
import joblib
from sklearn.metrics import accuracy_score
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LogisticRegression
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.metrics import classification_report, confusion_matrix, ConfusionMatrixDisplay

#Loading and exploring the Dataset
df = pd.read_csv("https://gist.githubusercontent.com/curran/a08a1080b88344b0c8a7/raw/0e7a9b0a5d22642a06d3d5b9bcbad9890c8ee534/iris.csv")
print(df.head())
print(df.shape)
print(df.columns)
print(df.info())
print(df.describe())


#Assigning feature and species
feature = df.drop("species", axis=1)
species = df["species"]

#Split the data
feature_train, feature_test, species_train, species_test = train_test_split(feature, species, test_size=0.2, random_state=42)

#Pre processing
scaler = StandardScaler()
feature_train = scaler.fit_transform(feature_train)
feature_test = scaler.transform(feature_test)

# Model Training and Testing
model = LogisticRegression(max_iter=200)
model.fit(feature_train, species_train)

species_pred = model.predict(feature_test)

#Confusion metrix
cm = confusion_matrix(species_test, species_pred)
disp = ConfusionMatrixDisplay(confusion_matrix=cm, display_labels=model.classes_)
disp.plot(cmap='Blues')
plt.show()

#Cecking Accuracy with Test Data
accuracy = accuracy_score(species_test, species_pred)
print(f"Accuracy: {accuracy:.4f}")

#Predicting the species based on new data
new_data = [[6.9, 3.1, 4.9, 1.5],[3, 3.1, 4.9, 1.5],[5.9, 3.1, 4.9, 1.5],[4.9, 2.7, 5.1, 2.5]]
data_scaled = scaler.transform(new_data)

predictions = model.predict(data_scaled)
print(predictions)

#Data Visualization
species = species.map({0: 'setosa', 1: 'versicolor', 2: 'virginica'})
sns.pairplot(df, hue='species', diag_kind='kde', palette='bright', height = 3)
plt.show()


#Summary
# Basic understanding of how to use different models based on algorithms
# Importance of Preprocessing i.e. without using the "StandardScaler" not able to achieve the desired accuracy
```
